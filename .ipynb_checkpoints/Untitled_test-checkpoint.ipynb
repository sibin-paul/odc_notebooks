```python
import datacube
from utils.dea_tools.plotting import rgb
```

```python
dc = datacube.Datacube(app = 'test_app')
```

```python
# latitude= (14.69,14.79)
# longitude = (74.51,74.65)

lat = 14.4595
lon = 74.4799

# lat = (73.79340127741148, 75.27948354279906)
# lon = (13.88058230140346, 15.154421893711945)

buffer = 0.05
product='ls8_c01_l2_usgs_sr_scene'
latitude = (lat - buffer, lat + buffer)
longitude = (lon - buffer, lon + buffer)
resolution = (-30, 30)
time_extents = ('2021-01-27')
platform="LANDSAT_8"

output_crs = 'EPSG:32643'
measurements = ['red', 'blue', 'green', 'nir', 'swir1', 'swir2', 'pixel_qa']
landsat_dataset = dc.load(product=product,
                          latitude=latitude,
                          longitude=longitude,
                          platform=platform,
                          time=time_extents,
                          output_crs=output_crs,
                          resolution=resolution,
                          measurements=measurements
)

landsat_dataset
```
